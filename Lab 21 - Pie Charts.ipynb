### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt


# Count the frequency of each database in 'DatabaseWantToWorkWith'
db_counts = df['DatabaseWantToWorkWith'].value_counts()

# Select the top 5 most desired databases
top5_dbs = db_counts.head(5)

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_dbs, labels=top5_dbs.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Databases Respondents Want to Work With")
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt


# Count the frequency of each developer type
dev_counts = df['DevType'].value_counts()

# Select the top 5 developer roles
top5_devs = dev_counts.head(5)

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_devs, labels=top5_devs.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Developer Roles Among Respondents")
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt


# Split multiple OS entries and flatten into a single series
all_os = df['OpSysProfessional use'].dropna().str.split(';').explode()

# Count frequency of each OS
os_counts = all_os.value_counts()

# Optionally, take the top 5 for clarity
top_os = os_counts.head(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=140)
plt.title("Top Operating Systems Used Professionally by Respondents")
plt.show()

```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Step 1: Drop missing values
df_languages = df.dropna(subset=['LanguageHaveWorkedWith'])

# Step 2: Split multiple languages (assuming comma-separated) into a list
all_languages = df_languages['LanguageHaveWorkedWith'].str.split(';').sum()

# Step 3: Count frequency of each language
language_counts = Counter(all_languages)

# Step 4: Get the top 5 most common languages
top5_languages = language_counts.most_common(5)
languages, counts = zip(*top5_languages)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=languages, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 Programming Languages Respondents Have Worked With')
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Step 1: Drop missing values
df_collab = df.dropna(subset=['NEWCollabToolsHaveWorkedWith'])

# Step 2: Split multiple tools (assuming semicolon-separated) into a list
all_tools = df_collab['NEWCollabToolsHaveWorkedWith'].str.split(';').sum()

# Step 3: Clean extra spaces
all_tools = [tool.strip() for tool in all_tools]

# Step 4: Count frequency of each tool
tool_counts = Counter(all_tools)

# Step 5: Get the top 5 most used tools
top5_tools = tool_counts.most_common(5)
tools, counts = zip(*top5_tools)

# Step 6: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=tools, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 Collaboration Tools Used Professionally')
plt.show()
```

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Step 1: Drop missing values
df_admired = df.dropna(subset=['LanguageAdmired'])

# Step 2: Split multiple languages if semicolon-separated
all_admired = df_admired['LanguageAdmired'].str.split(';').sum()

# Step 3: Clean extra spaces
all_admired = [lang.strip() for lang in all_admired]

# Step 4: Count frequency of each language
admired_counts = Counter(all_admired)

# Step 5: Get the top 5 most admired languages
top5_admired = admired_counts.most_common(5)
languages, counts = zip(*top5_admired)

# Step 6: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=languages, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top 5 Most Admired Programming Languages')
plt.show()
```
